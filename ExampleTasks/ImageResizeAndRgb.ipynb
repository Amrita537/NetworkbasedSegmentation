```python
import cv2
import numpy as np
from PIL import Image,ImageDraw,ImageColor,ImageFont
import matplotlib.pyplot as plt
import matplotlib
import webcolors
import os
#################### Global variables ##########################

file_name='0019_top_flat'

#################### Image file operations ##########################
def RGBImage(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    print("Image converted to RGB")
    return img

def ResizedImage(img):
    print("Received Image size ",img.shape[1],img.shape[0])
    scale_percent = 60
    width = int(img.shape[1] * scale_percent / 200)
    height = int(img.shape[0] * scale_percent / 200)
    dim = (width, height)
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    print("Image resized to ",dim)
    cv2.imwrite("resized/resized"+file_name+".png", np.hstack([resized]))
    print("Saving as resized"+file_name+".png")
    return resized
    
def show_image(img):
    plt.imshow(img)
    img = Image.fromarray(img)
    print("Image mode: ",img.mode)
    
def Get_ProcessedImage(img):
    img=RGBImage(img)
    img=ResizedImage(img)
    return img
#################### Image file operations ##########################

file_path ='sampleImages/'+file_name+'.png'
cv_img = cv2.imread(file_path)
img_file= Get_ProcessedImage(cv_img)
file_path2 ="resized/resized"+file_name+".png"

show_image(img_file)
```
